```python
import pandas as pd
import os
import ast
```

```python
folder_path = '/'.join(os.getcwd().split('/')[:-1]) + '/Clean Data/'
```

```python
a_tracks = pd.read_csv(folder_path + 'a_tracks.csv')
final_matrix = pd.read_csv(r'final_matrix_with_writers.csv')
```

```python
def list_transformation(line):
    return ast.literal_eval(line)
```

```python
a_tracks['producer'] = a_tracks['producer'].apply(lambda x: list_transformation(x))
a_tracks['writer'] = a_tracks['writer'].apply(lambda x: list_transformation(x))
```

```python
a_producers = []
a_writers = []

for item in a_tracks.iterrows():
    a_producers.extend(item[1].producer)
    a_writers.extend(item[1].writer)

print("{0} producers in total".format(len(a_producers)))
print("{0} writers in total".format(len(a_writers)))


a_producers = list(set(a_producers))
a_writers = list(set(a_writers))

print("{0} unique producers".format(len(a_producers)))
print("{0} unique writers".format(len(a_writers)))
```

```python
final_matrix['producer'] = final_matrix['producer'].apply(lambda x: '[]' if x == 'Connection Error' else x)
final_matrix['writer'] = final_matrix['writer'].apply(lambda x: '[]' if x == 'Connection Error' else x)

final_matrix['producer'] = final_matrix['producer'].apply(lambda x: list_transformation(x))
final_matrix['writer'] = final_matrix['writer'].apply(lambda x: list_transformation(x))
```

```python
final_matrix['missing_producer'] = final_matrix['producer'].apply(lambda x: 1 if len(x) == 0 else 0)
final_matrix['missing_writer'] = final_matrix['writer'].apply(lambda x: 1 if len(x) == 0 else 0)
```

```python
def check_a(line,role):
    if len(line) == 0:
        return 0, 0
    else:
        if role == 'producer':
            count = 0
            for l in line:
                if l in a_producers:
                    count += 1
                    return 1,count
                else:
                    return 0, count
        if role == 'writer':
            count = 0
            for l in line:
                if l in a_writers:
                    count += 1
                    return 1,count
                else:
                    return 0,count
                
    
```

```python
final_matrix['a_producer'] = final_matrix['producer'].apply(lambda x: check_a(x,'producer')[0])
final_matrix['a_producer_count'] = final_matrix['producer'].apply(lambda x: check_a(x,'producer')[1])

final_matrix['a_writer'] = final_matrix['writer'].apply(lambda x: check_a(x,'writer')[0])
final_matrix['a_writer_count'] = final_matrix['writer'].apply(lambda x: check_a(x,'writer')[1])
```

```python
final_matrix['producer_count'] = final_matrix['producer'].apply(lambda x: len(x))
final_matrix['writer_count'] = final_matrix['writer'].apply(lambda x: len(x))
```
